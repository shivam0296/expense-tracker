```python
# Importing Libraries

import sqlite3 as sql3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
conn = sql3.connect('expense_tracker.db')
```

```python
cursor = conn.cursor()
```

```python
file = '/Users/shivammishra/Downloads/expenses.csv'
df = pd.read_csv(file)
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# ignoring the null values
attributes = ['hrk','vendor', 'date', 'description','meansofpayment','city', 'category', 'currency','country']
new_data = df[attributes]
```

```python
new_data.rename(columns={'hrk': 'cost'}, inplace=True)
new_data
```

```python
# Creating the table 
sql_qry = """
create table expense(
id serial,
cost int not null,
vendor text,
date datetime,
meansofpayment text,
city varchar(50),
category varchar(10),
currency varchar(5),
counrty varchar(20)
)
"""

cursor.execute(sql_qry)
```

```python
# Inserting the data

new_data.to_sql('expense', conn, index=False, if_exists='replace')
```

```python
res = cursor.execute('select * from expense')
```
